## 1. Key governance themes for a scorecard

A bank-style scorecard framework usually needs more than model fit.

Typical questions are:
1. Are the variables stable and explainable?
2. Do bins behave monotonically?
3. Is the score population drifting?
4. Is calibration still appropriate?
5. Do approval cut-offs still align with risk appetite?
6. Has the underlying portfolio changed?
7. Do rejected applications create bias in observed outcomes?

## 10. Framework alignment — rating philosophy, retail pools, use test & independence

*(Added under the PD framework alignment: APS 113 / APG 113 / Basel CRE36 / WP14. This section
is **documentation**, not new modelling — for a portfolio demo these required framework elements
are stated rather than operationalised, but each names the rule it satisfies.)*

### 10.1 Asset class & rating dimension — these grades are retail **pools** (PD framework 2.4)

Home Credit is **consumer / retail** credit. Under the framework, retail exposures are not rated
one-by-one onto an obligor grade; they are assigned to **pools** of homogeneous risk. So the
**G1–G8 grades (and the A–E score bands) are best read as pools**, not corporate-style obligor
ratings. A production pool system would go further than this PD-only view and also:

- reflect **LGD and EAD** in the pool definition (not PD alone), and
- separate **delinquent vs current** exposures into different pools (a 30-days-past-due account and
  a clean account at the same score are not the same risk).

Here the pools are PD-ranked only; that boundary is intentional and stated as a limitation.

### 10.2 Default definition — `TARGET` is a competition **proxy** (APS 220 reference)

The modelled event is the Home Credit `TARGET` flag (a Kaggle competition label: serious
delinquency on the loan). The APS 220 reference definition of default is **unlikely-to-pay (UTP)
or 90+ days past due**. We treat `TARGET` as **broadly equivalent** to a 90+DPD/UTP default for
demonstration, but note two gaps: (i) it is a competition proxy, not a bank's booked default
register, and (ii) its **outcome horizon** is fixed by the competition, not set to a clean
12-month performance window. The companion EAD notebook (08) documents a *separate* data-quality
finding that the credit-card `SK_DPD` 90+ flag does not represent economic default; that finding
is about the card file, and does not change the application-scorecard `TARGET` used here.

### 10.3 Rating philosophy — point-in-time-leaning, long-run **calibrated** (APG 113 para 73)

This is an **application** scorecard built on origination-time features (bureau, external scores,
employment, demographics). It is therefore **point-in-time-leaning**: the *rank ordering* moves
with the applicant's current risk, not a through-the-cycle average. We approximate a
through-the-cycle **level** by calibrating the grade PDs to a long-run / target default rate
(see notebook 04 and `calibration_summary.csv`).

> **APG 113 para 73 caveat (stated explicitly):** calibrating point-in-time ratings to a long-run
> average **does not by itself make the ratings through-the-cycle.** The *level* is anchored long-run;
> the *dynamics* remain PIT. A genuinely TTC rating would require cycle-spanning data and a stressed
> or hybrid design, which a single-snapshot dataset cannot support.

### 10.4 Use test, overrides & independence (Part 4.1 / Step 8 / Part 5.8)

| Framework element | Status in this project | What it would be in production |
|---|---|---|
| **Use test** (Part 4.1) | This is a **demonstration** model, **not** used in live credit approval or pricing. | The same PDs would drive approval cut-offs, risk-based pricing, limits, provisioning and capital — and that use would be *evidenced*, not assumed. |
| **Overrides** (Step 8) | **None applied.** Section 6 above sketches the policy. | A written override policy with approval roles, separate up/down override tracking, and regular override-rate reporting by grade. |
| **Independence** (Part 5.8) | Development and validation are separated only **by notebook** (03 builds, 04 validates). | Independent **model development** and **model validation** functions, with validation reporting to model risk governance, not to the developer. |

### 10.5 Reject inference (development-sample bias)

The model is trained on the **accepts-only** booked population — rejected applicants have no
realised outcome. This biases the observed relationships toward the through-the-door population
that was actually approved. **No reject inference is applied** here. In production, reject inference
(e.g. parcelling, augmentation, or a bureau-outcome bring-back) would be required before the
scorecard could be claimed to represent the full applicant population.

### 10.6 What was added for framework alignment (cross-reference)

- **5 bps PD floor** — `src/calibration.py: apply_pd_floor` (APS 113 Att B para 1). Min grade PD
  is ~6–11%, so the floor does not bind; applied for completeness.
- **Calibration test** — `outputs/tables/scorecard_outputs/13_calibration_test.csv`: per-grade one-sided
  **binomial** under-estimation test + an overall **Hosmer-Lemeshow** statistic with a
  green/amber/red flag (PD framework Part 5.3). *Independence caveat (WP14): both tests assume
  uncorrelated defaults and therefore understate Type-I error when defaults are correlated, so
  amber/red flags are review prompts, not hard pass/fail.* On the calibrated PD every grade passes
  the binomial under-estimation test; Hosmer-Lemeshow flags that the calibrated **level** sits
  above the benign test-window default rate (conservative, by design of the long-run anchor).
- **Margin of conservatism** — `outputs/tables/scorecard_outputs/14_pd_moc_overlay.csv`: a documented
  additive +15% overlay on each grade PD (CRE36.67 / Step 10), sized for the proxy default
  definition, the absence of through-the-cycle data, and the accepts-only sample.
