## 2. Backtesting

```
# Example backtesting (requires time column)
DATE_COL = None  # replace with your date column

def backtest_default_rate(df, date_col, target_col="TARGET"):
    df[date_col] = pd.to_datetime(df[date_col])
    summary = df.groupby(df[date_col].dt.to_period("M"))[target_col].mean().reset_index()
    summary.columns = ["month", "default_rate"]
    return summary

print("Set DATE_COL to run backtesting")
```
